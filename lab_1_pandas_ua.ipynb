### 9a

Побудуйте графік розсіювання на вісях `age` та `Debt`. Cинім відміттє клієнтів без серйозних заборгованостей (`SeriousDlqin2yrs = 0`) та червоним - боржників (`SeriousDlqin2yrs = 1`).

```python
# Зчитуємо дані з файлу CSV
df = pd.read_csv('/content/sample_data/data.csv', delimiter=',')

# Розділяємо дані на дві підмножини за значенням SeriousDlqin2yrs
non_defaulters = df[df['SeriousDlqin2yrs'] == 0]
defaulters = df[df['SeriousDlqin2yrs'] == 1]

# Побудова графіку розсіювання для кожної підмножини
plt.scatter(non_defaulters['age'], non_defaulters['MonthlyIncome'], c='blue', label='Без заборгованості')
plt.scatter(defaulters['age'], defaulters['MonthlyIncome'], c='red', label='З заборгованістю')

# Назви вісей та заголовок
plt.xlabel('Вік')
plt.ylabel('Місячний дохід')
plt.title('Графік розсіювання між віком та місячним доходом')

# Легенда
plt.legend()

# Показати графік
plt.show()
```

### 9b
Побудуйте на однрму графіку дві **нормовані** щільності розподілення: червону – для місяіного доходу клієнтів з заборгованостями, синю - для місяіного доходу клієнтів без заборгованостей. По вісі абсцис відобразіть значення до 25000.

```python
# Зчитуємо дані з файлу CSV
df = pd.read_csv('/content/sample_data/data.csv', delimiter=',')

# Розділяємо дані на дві підмножини за значенням SeriousDlqin2yrs
non_defaulters = df[df['SeriousDlqin2yrs'] == 0]
defaulters = df[df['SeriousDlqin2yrs'] == 1]

# Побудова графіку нормованих щільностей розподілення місячного доходу
plt.hist(non_defaulters['MonthlyIncome'], bins=50, density=True, color='blue', alpha=0.5, label='Без заборгованості')
plt.hist(defaulters['MonthlyIncome'], bins=50, density=True, color='red', alpha=0.5, label='З заборгованістю')

# Обмеження області графіку за віссю X
plt.xlim(0, 25000)

# Назва вісі абсцис та заголовок
plt.xlabel('Місячний дохід')
plt.title('Нормовані щільності розподілення місячного доходу')

# Легенда
plt.legend()

# Показати графік
plt.show()
```

Візуалізуйте попарні залежності між небінарними ознаками попарные зависимости между небинарными признаками `'age', 'MonthlyIncome', 'NumberOfDependents'`. Обмежте при цьому місячний дохід значенням 25000.

Які закономірності ви можете спостерігати на отриманих графіках?

*Функції, що можуть бути корисними при вирішенні: `pd.plotting.scatter_matrix()`*

```python
df = pd.read_csv('/content/sample_data/data.csv', delimiter=',')

# Обмеження місячного доходу значенням 25000
df = df[df['MonthlyIncome'] <= 25000]

# Вибираємо потрібні ознаки для аналізу
selected_features = ['age', 'MonthlyIncome', 'NumberOfDependents']
df_selected = df[selected_features]

# Побудова матриці графіків розсіювання
scatter_matrix = pd.plotting.scatter_matrix(df_selected, figsize=(10, 10), diagonal='hist', alpha=0.5)

# Назви вісей та заголовок
plt.suptitle('Попарні залежності між ознаками')

# Відображення графіків
plt.show()
```
